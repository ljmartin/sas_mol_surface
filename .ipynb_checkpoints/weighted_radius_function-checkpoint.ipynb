```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.spatial.distance import cdist
from scipy.special import expit
```

```python
#create a grid of the correct size:
grid_size = 100
a,b = np.ones([grid_size, grid_size]).nonzero()
gPoints = np.vstack([a,b]).T
print(gPoints.shape)
```

```python
centres = np.array([[45,50], [20,20]])
radii = np.array([30,20])
```

```python
#now do some function to 'm' that weights between distances. 
k = 0.1
def squared_exponential(distance):
    return np.exp(distance**2/ -10)

def sigmoid(x):
    return 1-expit(x/2-4)

x = np.linspace(-1,15, 100)
#plt.plot(x, squared_exponential(x))
plt.plot(x, sigmoid(x))
plt.grid()
```

```python
distances =cdist(gPoints, centres)
```

```python
argpart = np.argpartition(distances-radii, 1, axis=1)
one_nn = argpart[:,0]
#all_others = argpart[:,1:]
```
